## Importing the libraries

```python
from fastai.text.all import *
import torch
```

## Loading the IMDB dataset

```python
path = untar_data(URLs.IMDB)
```

## Building the Classification Datablock

`TextBlock` and `CategoryBlock` is used to handle text and label data.

```python
dls_clas = DataBlock(
    blocks=(TextBlock.from_folder(path), CategoryBlock),
    get_y=parent_label,
    get_items=partial(get_text_files, folders=['train', 'test']),
    splitter=GrandparentSplitter(valid_name='test')
).dataloaders(path, bs=128, seq_len=72)
```

## Creating the text classifier learner

Creating a learner using AWD_LSTM model and the created DataLoaders.

```python
learn = text_classifier_learner(dls_clas, AWD_LSTM, drop_mult=0.5, metrics=accuracy).to_fp16()
```

## Training the model

Classifier is fine-tuned.
Fine-tuning the model for 1 epoch with a learning rate of 0.02

```python
learn.fine_tune(1, 2e-2)
```

## Evaluating the model

The model is evaluated using the validation set.

```python
learn.show_results()
```

## Model is tested using sample text

```python
test_review1 = "The movie could have been better as it lacked a proper story and the ending was confusing."
test_review2 = "I will certainly remember this movie as it moved me and I also cried."

# using predict the model is tested

# For test_review1 which is negative experience
pred, pred_idx, probs = learn.predict(test_review1)
print(f"\nPredicted Sentiment: {pred}")
print(f"Prediction Probabilities: {probs}")

# For test_review2 which is positive experience
pred, pred_idx, probs = learn.predict(test_review2)
print(f"\nPredicted Sentiment: {pred}")
print(f"Prediction Probabilities: {probs}")
```
